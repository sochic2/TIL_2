```python
def convert_markdown_table(table):
    """
    table 내용이 담긴 2중 list를 받아 마크다운 테이블 형식으로 return

    :param table: table 내용이 담긴 2중
    :return: 마크다운 테이블 포맷의 txt
    """
    rows = []
    for idx, row in enumerate(table):
        rows.append('| ' + ' | '.join(map(str, row)).replace('\n', '').replace('None', '') + ' |')
        if idx == 0:
            rows.append('| ' + ' | '.join(['---'] * len(row)) + ' |')
    return '\n' + '\n'.join(rows) + '\n'
```

```python
import pdfplumber

def group_words_by_rows(words, y_tolerance=5):
    rows = []
    current_row = []
    current_y = None

    # 단어들을 Y축 기준으로 정렬
    words = sorted(words, key=lambda w: (w['top'], w['x0']))

    for word in words:
        y_top = word['top']

        # 새로운 행 생성 조건 (y_tolerance 사용)
        if current_y is None or abs(current_y - y_top) > y_tolerance:
            if current_row:
                rows.append(sorted(current_row, key=lambda w: w['x0']))
            current_row = [word]
            current_y = y_top
        else:
            current_row.append(word)

    # 마지막 행 추가
    if current_row:
        rows.append(sorted(current_row, key=lambda w: w['x0']))

    return rows

```

```python
# PDF 파일 열기
with pdfplumber.open("02-용지별 시행지침(1-1,1-3시행지침)(변경)_.pdf") as pdf:
    page = pdf.pages[0]
    tables = page.find_tables()
    table = tables[0]
    within_page = page.within_bbox(table.bbox)

    # 페이지에서 단어와 좌표 추출
    words = within_page.extract_words()

    # 단어를 기반으로 테이블 형성
    table_rows = group_words_by_rows(words)

    # 테이블 출력
    for row in table_rows:
        print("\t".join([word['text'] for word in row]))
```
